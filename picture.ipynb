```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a sphere
r = 1
pi = np.pi
cos = np.cos
sin = np.sin
phi, theta = np.mgrid[0.0:pi:50j, 0.0:2.0*pi:50j]
x = r*sin(phi)*cos(theta)
y = r*sin(phi)*sin(theta)
z = r*cos(phi)

# Create a 3D plot
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection='3d')

# Plot the sphere
ax.plot_surface(x, y, z, color='b', alpha=0.1)

# Plot the 3D coordinate axis
l = 1.5*r
ax.plot([-l, l], [0, 0], [0, 0], color='k', linestyle='--', alpha=0.5)
ax.plot([0, 0], [-l, l], [0, 0], color='k', linestyle='--', alpha=0.5)
ax.plot([0, 0], [0, 0], [-l, l], color='k', linestyle='--', alpha=0.5)


# ax.quiver(0, 0, 0, 0, 0, 1,  
#  length=1,normalize=True,linewidths=5,colors='green',arrow_length_ratio=0.1)

# ax.quiver(0, 0, 0, -0.5, 0, 1,  
#  length=1,normalize=True,linewidths=5,arrow_length_ratio=0.1)

# ax.quiver(0, 0, 0, 1, 0, 0,  
#  length=1,normalize=True,linewidths=5,arrow_length_ratio=0.1)

v1 = np.array([0, 0, 1])
v2 = np.array([-0.5, 0.5, 0])
v3 = np.array([0.5, 0, 0.5])

# Calculate the cosine similarity of each pair of vectors
cos_sim12 = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
cos_sim13 = np.dot(v1, v3) / (np.linalg.norm(v1) * np.linalg.norm(v3))
cos_sim23 = np.dot(v2, v3) / (np.linalg.norm(v2) * np.linalg.norm(v3))

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='g')
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='b')

# Annotate the cosine similarity of each pair of vectors
ax.text(v1[0]*1.1, v1[1]*1.1, v1[2]*1.1, f'{cos_sim12:.2f}', color='r')
ax.text(v1[0]*1.1, v1[1]*1.1, v1[2]*0.9, f'{cos_sim13:.2f}', color='r')
ax.text(v2[0]*1.1, v2[1]*1.1, v2[2]*0.9, f'{cos_sim23:.2f}', color='g')



# Set the plot limits and labels
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
